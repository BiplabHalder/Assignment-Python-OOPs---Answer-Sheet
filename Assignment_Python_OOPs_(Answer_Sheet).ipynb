#Python OOPs Questions
 1. What is Object-oriented Programming (OOP)?
  - Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects",
which contain data in the form of fields (attributes) and code in the form of methods.
It helps in organizing complex programs, promotes code reusability, and improves maintainability.
  

 2. What is a class in OOP?
  - A class is a blueprint for creating objects. It defines a set of attributes and methods
that the created objects (instances) will have.
  

 3. What is an object in OOP?
  - An object is an instance of a class. It represents a specific implementation of the class
with actual values and can use the class methods and attributes.

 4. What is the difference between abstraction and encapsulation?
  - Abstraction hides complexity by showing only essential details. Encapsulation binds data and
methods together and restricts direct access to some of the objectâ€™s components.

 5. What are dunder methods in Python?
  - Dunder (double underscore) methods are special methods in Python that start and end with double underscores,
like __init__, __str__, __len__, etc. They are used to define object behavior for built-in operations.

 6. Explain the concept of inheritance in OOP.
  - Inheritance allows a class (child) to acquire properties and methods from another class (parent).
It promotes code reusability and hierarchical relationships.

 7. What is polymorphism in OOP?
  - Polymorphism means the same function or method can behave differently on different classes or data types.
It allows different classes to define methods that are called through the same name.

 8. How is encapsulation achieved in Python?
  - Encapsulation is achieved using private (single or double underscore) variables and providing
public methods (getters/setters) to access or modify them.

 9. What is a constructor in Python?
  - A constructor is a special method (__init__) in a class that is automatically called
when an object is created. It initializes the object's attributes.

 10. What are class and static methods in Python?
  -  -  Class methods use @classmethod decorator and take 'cls' as the first argument. They can access or modify class state.
      -  Static methods use @staticmethod and do not take 'self' or 'cls'. They behave like regular functions within a class.
  
11. What is method overloading in Python?
  - Python does not support traditional method overloading. However, it can be achieved using default arguments or *args/**kwargs.

 12. What is method overriding in OOP?
  - Method overriding allows a subclass to provide a specific implementation of a method already defined in its superclass.

 13. What is a property decorator in Python?
  - @property is a built-in decorator used to define getter methods. It allows a method to be accessed like an attribute.

 14. Why is polymorphism important in OOP?
  - Polymorphism increases flexibility and maintainability of code by allowing objects of different classes
to be treated through a common interface.

 15. What is an abstract class in Python?
  - An abstract class is a class that cannot be instantiated directly. It can contain abstract methods
that must be implemented by its child classes. It is defined using the 'abc' module.

 16. What are the advantages of OOP?
  -     - Code reusability
        -  Scalability
        -  Maintainability
        -  Better organization and structure
        -  Real-world modeling


 17. What is the difference between a class variable and an instance variable?
-     - Class variables are shared across all instances of a class.
      - Instance variables are unique to each object.

 18. What is multiple inheritance in Python?
  - Multiple inheritance means a class can inherit from more than one parent class.
Example: class C(A, B)


 19. Explain the purpose of '__str__' and '__repr__' methods in Python.
  -     -__str__ returns a readable string representation of an object.
       - __repr__ returns an unambiguous string useful for debugging and development.

 20. What is the significance of the 'super()' function in Python?
  -  super() is used to call methods from the parent class. It is commonly used in method overriding and constructors.

 21. What is the significance of the '__del__' method in Python?
  - __del__ is a destructor method that is called when an object is about to be destroyed.
It is used to clean up resources.
  
22. What is the difference between @staticmethod and @classmethod in Python?
  -    - @staticmethod does not access class or instance variables.
       - @classmethod takes the class as its first argument and can modify class state.

 23. How does polymorphism work in Python with inheritance?
  - In inheritance, a subclass can override methods of the parent class, and Python can call the appropriate method
based on the object type at runtime.

 24. What is method chaining in Python OOP?
  - Method chaining is the practice of calling multiple methods on the same object in a single line.
Each method returns 'self' to continue the chain.

 25. What is the purpose of the '__call__' method in Python?
  - __call__ allows an instance of a class to be called as if it were a function. It makes objects callable.

    def area(self):

5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance.
-   Encapsulation in BankAccount

class BankAccount:

    def __init__(self):

        self.__balance = 0


    def deposit(self, amount):

        self.__balance += amount


    def withdraw(self, amount):

        if amount <= self.__balance:

            self.__balance -= amount

        else:

            print("Insufficient balance")

    def check_balance(self):

        return self.__balance

-	 Output

print("5. Output:")

account = BankAccount()

account.deposit(1000)

account.withdraw(300)
